```python
class Node():
    
    def __init__(self, value):
        self.value = value 
        self.next = None

    
class LinkedList():
    
    def __init__(self):
        self.first = None
        self.last = None
        self.size = 0
    
    
    def isEmpty(self):
        return self.size == 0
    
    
    def displayContent(self):
        
        current = self.first
        content = []
        while current:
            content.append(current.value)
            current = current.next
        
        print(content)
            
            
    def addFirst(self, value):
        new_node = Node(value)
        
        if self.isEmpty():
            self.first = self.last = new_node
        else:
            new_node.next = self.first
            self.first = new_node
        
        self.size += 1
        
        
    def addLast(self, value):
        new_node = Node(value)
    
        if self.isEmpty():
            self.first = self.last = new_node
        else:
            self.last.next = new_node
            self.last = new_node 
        
        self.size += 1
    
    
    def removeFirst(self):
        
        if self.isEmpty():
            return 
        else:
            second = self.first.next 
            self.first.next = None
            self.first = second
        
        self.size -= 1
    
    
    def findPreviousNode(self, node):
        
        current = self.first 
        while current:
            if (current.next == node):
                return current
            current = current.next
        
        return None
        
        
    def removeLast(self):
        
        if self.isEmpty():
            return 
        else:
            previous = self.findPreviousNode(self.last)
            self.last = previous
            self.last.next = None
            
        self.size -= 1
    
    
    def indexOf(self, value):
        
        index = 0
        current = self.first
        while current: # while current is not equal to zero
            if (current.value == value):
                return index
            current = current.next
            index += 1
        
        return -1
    
    
    def contains(self, value):
        
        current = self.first
        while current:
            if (current.value == value):
                return True
            current = current.next
        
        return False
    
    
    def reverse(self):
        
        if self.isEmpty():
            return 
        
        previous = self.first
        current = self.first.next
        
        while current:
            second = current.next
            current.next = previous
            previous = current
            current = second 
        
        self.last = self.first
        self.last.next = None
        self.first = previous
        
    
    def getKthFromTheEnd(self, k):
        
        if self.isEmpty():
            return 
        
        first_pointer = self.first
        second_pointer = self.first
        distance = k
        
        if self.size < k:
            print("Invalid Input. K must be smaller than list.")
            return 
        
        for _ in range(0, distance):
            second_pointer = second_pointer.next
        
        while second_pointer.next:
            second_pointer = second_pointer.next
            first_pointer = first_pointer.next
        
        return first_pointer.value
```

```python
ll = LinkedList()

ll.addFirst(10)
ll.addFirst(20)
ll.addFirst(30)
ll.addFirst(40)

ll.addLast(50)

# ll.removeFirst()

ll.removeLast()

print(ll.indexOf(20))

print(ll.contains(10))

print(ll.size)

ll.displayContent()

print(ll.reverse())

print(ll.getKthFromTheEnd(2))
```
