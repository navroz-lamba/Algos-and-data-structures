```python
def csBSTRangeSum(root, lower, upper):
    if (root == None):
        return 0
    
    if root.value > upper:
        return csBSTRangeSum(root.left, lower, upper)
    
    elif root.value < lower:
        return csBSTRangeSum(root.right, lower, upper)
    
    else:
        return csBSTRangeSum(root.left, lower, upper) + root.value + csBSTRangeSum(root.right, lower, upper)
```
